```python
# 3. Analyzing the 'churn' variable 
data.head()
```

```python
# unique()
data['churn'].unique()
```

```python
# value_counts()
A=data['churn'].value_counts()
A
```

```python
# if we check the type A then we use simple function below
type(A)
```

```python
# Now starts visualization of this column 
# 1. using Donut chart - it is simply modified piechart, within an area from center cut out , now first draw a pieplot using matplotlib library
# plt.pie(A)
# plt.pie(A, labels =['Not churned','churned'])
# plt.pie(A, labels =['Not churned','churned'], colors = ['Orange', 'Lime'])
# plt.pie(A, labels =['Not churned','churned'], colors = ['Orange', 'Lime'], startangle=50)
plt.pie(A, labels =['Not churned','churned'], colors = ['Orange', 'Lime'], startangle=50, shadow=True,radius=2,explode=(0,0.2), autopct='%1.2f%%',pctdistance=0.75)
# 2. using plt.circle we will create a circle and save it in the variable
circle = plt.Circle((0,0),1,color='white') 
c =plt.gcf()
# 3. we will add circle at centre of piechart using gca().add_artist()
c.gca().add_artist(circle)
plt.title("Churned vs not curned customers(DSL)")
plt.show()
```
